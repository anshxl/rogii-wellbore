# Phase 2 EDA — per-well diagnosis of the worst 20 wells

Renders 3-panel diagnostic plots for the top-20 worst wells (by ensemble OOF RMSE) plus 5 easy contrast wells. Output figures land in `eda_outputs/figs/phase2/`.

Spec: [docs/superpowers/specs/2026-05-08-phase2-eda-worst-wells-design.md](../docs/superpowers/specs/2026-05-08-phase2-eda-worst-wells-design.md)

```python
%load_ext autoreload
%autoreload 2
import sys
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

ROOT = Path('..').resolve()
sys.path.insert(0, str(ROOT / 'src'))
import os
os.chdir(ROOT)

from eda_phase2_plots import render_wells
```

```python
per_well = pd.read_parquet('artefacts/per_well_oof.parquet')

WORST_20 = per_well.sort_values('rmse_ens', ascending=False).head(20)['well'].tolist()
median = per_well['rmse_ens'].median()
EASY_5 = (
    per_well.iloc[(per_well['rmse_ens'] - median).abs().argsort()[:5]]
    .sort_values('rmse_ens')['well'].tolist()
)

print('worst 20:', WORST_20)
print('easy 5:', EASY_5)
```

```python
all_wells = WORST_20 + EASY_5
paths = render_wells(all_wells)
print(f'rendered {len(paths)} figures')
```

## Worst 20

```python
for w in WORST_20:
    print(w)
    display(Image(filename=f'eda_outputs/figs/phase2/{w}.png'))
```
